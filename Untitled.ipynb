```python
import csv
from collections import defaultdict

def sum_marks(input_filename, output_filename):
    marks_dict = defaultdict(int)


    try:
        with open(input_filename, mode='r') as infile:
            reader = csv.reader(infile)
            next(reader)
            for row in reader:
                roll_num, marks = row
                marks_dict[roll_num] += int(marks)
    except FileNotFoundError:
        print(f"Error: The file {input_filename} was not found.")
        return

    with open(output_filename, mode='w', newline='') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(['Roll Num', 'Marks'])
        for roll_num, total_marks in marks_dict.items():
            writer.writerow([roll_num, total_marks])

    print(f'Output written to {output_filename}')

input_filename = 'Mini Project - Marks Adding.csv'
output_filename = 'output.csv'

sum_marks(input_filename, output_filename)
```
